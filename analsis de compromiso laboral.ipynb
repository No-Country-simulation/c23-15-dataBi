#**Compromiso en las modalidades de trabajo**: ¿Cómo afecta el trabajo remoto o híbrido al desempeño y compromiso de los empleados?

En los últimos años, el trabajo remoto ha transformado la dinámica laboral a nivel global, impulsando cambios en la gestión del talento, la productividad y la cultura organizacional. Si bien esta modalidad ha permitido mayor flexibilidad, también ha planteado interrogantes sobre su impacto en el desempeño, la satisfacción y la retención del talento.

Este análisis explora la relación entre la frecuencia de viajes y factores laborales clave, como el rendimiento, el bienestar y la rotación de empleados. Además, examina cómo las modalidades de trabajo remoto e híbrido influyen en estas dinámicas, considerando diferencias entre departamentos, roles y características demográficas.

Al desentrañar estas interacciones, este estudio busca aportar insights estratégicos para optimizar el compromiso y la eficiencia en entornos de trabajo flexibles, permitiendo a las organizaciones diseñar políticas más alineadas con las necesidades de sus colaboradores y objetivos operativos.

### **Demografico**

#### Cantidad de viajes segun su estado civil

```python
df_maried_status = employee_df.groupby(['MaritalStatus','BusinessTravel'])['BusinessTravel'].count().reset_index(name ='Count')
# Sumar los valores de 'Count' para cada categoría de 'BusinessTravel'
sum_counts = df_maried_status.groupby('BusinessTravel')['Count'].sum().reset_index()

# Calcular el porcentaje específico para cada categoría
sum_counts['Percentage'] = (sum_counts['Count'] / sum_counts['Count'].sum() * 100).round(2)

# Crear un DataFrame para almacenar los resultados finales
resultados_finales = pd.DataFrame()

# Iterar sobre cada categoría y calcular el porcentaje para cada usuario
for categoria in sum_counts['BusinessTravel']:
    # Filtrar filas para la categoría actual
    filtered_maried_status = df_maried_status.loc[df_maried_status['BusinessTravel'] == categoria]

    # Calcular el porcentaje para cada usuario en la categoría actual
    filtered_maried_status['bt_Percentage'] = (filtered_maried_status['Count'] / sum_counts.loc[sum_counts['BusinessTravel'] == categoria, 'Count'].values[0] * 100).round(2)

    # Añadir los resultados al DataFrame final
    resultados_finales = pd.concat([resultados_finales, filtered_maried_status])

resultados_finales
```

```python
# Graficar una barra apilada
plt.figure(figsize=(10, 6))

# Pivotar el DataFrame para obtener el formato adecuado para el gráfico apilado
pivot_df = resultados_finales.pivot(index='MaritalStatus', columns='BusinessTravel', values='bt_Percentage')
colors = ['#5C23F5', '#C673F5', '#8A2BE2']

# Graficar las barras apiladas para cada categoría de BusinessTravel
ax = pivot_df.plot(kind='bar', stacked=True, color=colors)

# Personalizar el gráfico
plt.title("Distribución de la Frecuencia de Viajes según Categoría")
plt.xlabel("Estado civil")
plt.ylabel("Porcentaje (%)")
plt.xticks(rotation=0)
plt.legend(title="Categoría de Viaje", loc='upper left')

# Agregar los porcentajes en cada sección de la barra
for p in ax.patches:
    height = p.get_height()
    width = p.get_width()
    x = p.get_x() + width / 2
    y = p.get_y() + height / 2

    if height > 0:
        ax.text(x, y, f'{height:.1f}%', ha='center', va='center', fontsize=10, color='black')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

**Conclusión**:
- Existe una mayor cantidad de emplados casados donde se mantiene cosistente como los empleados que no viajan son la minoria

### ***Responsabilidades laborales***

#### *Frecuencia de viaje por departamento*

```python
Departament_Frecuency_Travel = employee_df.groupby(['Department','BusinessTravel'])['DailyRate'].count().reset_index()
Frecuency_Travel = Departament_Frecuency_Travel.groupby('BusinessTravel')['DailyRate'].sum().reset_index()

Departament_Frecuency_Travel.columns = ['Department', 'BusinessTravel', 'Frecuency']
Frecuency_Travel.columns = ['BusinessTravel', 'Frecuency']

Departament_Frecuency_Travel['Frecuency (%)'] = ((Departament_Frecuency_Travel['Frecuency'] / Departament_Frecuency_Travel['Frecuency'].sum())*100).round(2)
Frecuency_Travel['Frecuency (%)'] = ((Frecuency_Travel['Frecuency'] / Frecuency_Travel['Frecuency'].sum())*100).round(2)

display(Departament_Frecuency_Travel)
display(Frecuency_Travel)
```

```python
# Cambiar nombres y colores
labels = ['Inexistente', 'Frecuente', 'Ocacionalmente']  # Nuevos nombres
colors = ['#A480F5', '#3273F5', '#ADE1F5']  # Colores personalizados

# Crear el gráfico de pastel
plt.figure(figsize=(6, 6))
sns.set_style("whitegrid")
plt.pie(
    Frecuency_Travel['Frecuency (%)'],
    labels=labels,  # Usar nuevos nombres
    autopct='%1.1f%%',
    colors=colors  # Usar colores personalizados
)

# Agregar un título
plt.title("Frecuencia de viajes")

# Mostrar el gráfico
plt.show()
```

```python
# Configurar el tamaño de la figura
plt.figure(figsize=(10, 8))

# Reorganizar los datos para acumulación
departments = Departament_Frecuency_Travel['Department'].unique()
categories = Departament_Frecuency_Travel['BusinessTravel'].unique()
colors = ['darkblue', 'lightblue', '#AD38F5']

# Inicializar la base acumulada (inicialmente cero)
bottom = [0] * len(departments)

# Graficar cada categoría de forma acumulativa
for i, category in enumerate(categories):
    category_data = Departament_Frecuency_Travel[Departament_Frecuency_Travel['BusinessTravel'] == category]

    # Extraer los valores de frecuencia para cada departamento
    values = [category_data[category_data['Department'] == dept]['Frecuency (%)'].sum() for dept in departments]

    # Crear la barra acumulada
    plt.bar(
        departments,
        values,
        bottom=bottom,  # Apilar sobre la base acumulada
        color=colors[i],
        label=category
    )

    # Agregar los valores encima de cada barra
    for j, value in enumerate(values):
        if value > 0:  # Solo mostrar valores mayores a 0
            plt.text(
                x=j,  # Posición en el eje x
                y=bottom[j] + value / 2,  # Posición en el eje y (mitad de la sección actual)
                s=f"{value:.2f}%",  # Formato del porcentaje
                ha='center',  # Alineación horizontal
                va='center',  # Alineación vertical
                fontsize=10,  # Tamaño de la fuente
                color='white' if i < 2 else 'black'  # Contraste de color según el fondo
            )

    # Actualizar la base acumulada
    bottom = [bottom[j] + values[j] for j in range(len(departments))]

# Agregar la leyenda
plt.legend(title="Business Travel", loc='upper right')

# Agregar título y etiquetas
plt.title("Frecuencia de Viajes por Departamento")
plt.xlabel("Departamento")
plt.ylabel("Frecuencia (%)")

# Mostrar el gráfico
plt.show()
```

#### Balance entre la vida y el trabajo

```python
df_WorkLifeBalance = employee_df.groupby(['WorkLifeBalance','BusinessTravel'])['BusinessTravel'].count().unstack()
df_WorkLifeBalance
```

```python
# Calcular el total de viajes por género
total_balance_life = df_WorkLifeBalance.sum(axis=1)

# Calcular el total de viajes en general
total_balance = df_WorkLifeBalance.sum().sum()

# Calcular el porcentaje para cada categoría de BusinessTravel por separado y en conjunto
percentage_by_balance_life = df_WorkLifeBalance.divide(total_balance_life, axis=0) * 100
percentage_all_balance_life = df_WorkLifeBalance.divide(total_balance) * 100

# Mostrar los resultados
print("Porcentaje por género:")
print(percentage_by_balance_life)

print("\nPorcentaje conjunto (total de viajes):")
print(percentage_all_balance_life)
```

```python
# Datos
departments = percentage_all_balance_life.index
categories = percentage_all_balance_life.columns
colors = ['darkblue', 'lightblue', '#AD38F5']

# Configurar el tamaño de la figura
plt.figure(figsize=(7, 3))

# Inicializar la base acumulada (inicialmente cero)
bottom = np.zeros(len(departments))

# Graficar cada categoría de forma acumulativa
for i, category in enumerate(categories):
    values = percentage_all_balance_life[category].values

    # Crear la barra acumulada
    plt.barh(
        departments,
        values,
        left=bottom,  # Apilar sobre la base acumulada
        color=colors[i],
        label=category
    )

    # Actualizar la base acumulada
    bottom += values

# Agregar la leyenda
plt.legend(title="Business Travel", loc='lower right')

# Agregar título y etiquetas
plt.title("Frecuencia de Viajes por Equilibrio entre Vida y Trabajo")
plt.xlabel("Frecuencia (%)")
plt.ylabel("WorkLifeBalance")

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

Conclusión:

**Frecuencia de Viajes por Equilibrio entre Vida y Trabajo (Work-Life Balance):**
* La mayoría de los empleados con un balance vida-trabajo alto (Work-Life Balance = 3) tienden a viajar ocasionalmente (Travel_Rarely), lo cual representa un equilibrio aceptable entre las demandas laborales y personales.
* Los empleados con balance vida-trabajo bajo (Work-Life Balance = 1 o 2) tienen una menor proporción de viajes ocasionales y se distribuyen entre aquellos que no viajan (Non-Travel) o viajan frecuentemente (Travel_Frequently). Esto sugiere que viajar con demasiada frecuencia puede afectar negativamente el balance vida-trabajo.
